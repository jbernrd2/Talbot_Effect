```python
def count_boxes(low, high, left, right, epsilon, func, inputs):
    
    ind_sets = set()
    for dat in inputs:
        val = func.subs(x, dat).evalf()
#         print(val)
        ind_left, ind_up = (val - low)//epsilon, (dat - left)//epsilon
#         print(ind_left, ind_up)
        if ind_left > (right - left)//epsilon or ind_left < 0 or ind_up > (high - low)//epsilon or ind_up < 0:
            continue
            
        else:
            ind_sets.add((ind_left, ind_up))

    return ind_sets
```

```python
result = count_boxes(0, 1, 0, 1, 0.01, a, np.linspace(0,1,100))
# Fourier Analysis 
```
